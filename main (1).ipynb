```python
# A continuación, vamos a crear una columna nueva para source y a introducir usando regex los valores contenidos en 
#Investigator or Source
shark3['Investigator or Source']= shark3['Investigator or Source'].astype(str) #Pasamos el tipo a string.
shark3['Source'] = shark3.loc[:, 'Investigator or Source'] #Creamos una columna nueva a partir de la de investigadores

#Importamos Regex (re)
import re

# realizamos el código de regex  r'\b[A-Z][.| ][ |.\ [A-Z]\S+'
regex= r'\b[A-Z][.| ][ |.\ [A-Z]\S+'
```

```python
regex= r'\b[A-Z][.| ][ |.\ [A-Z]\S+'

shark3['Source'].replace(to_replace=regex, value=' ', regex=True, inplace=True)
#Ahora ejecutamos el regex reemplazando las coincidencias.
```

```python
shark3['Source'].head(30) #comprobamos que regex ha funcionando.
```

```python
def sharkre(string):
    return re.findall(r'\b[A-Z][.| ][ |.\ [A-Z]\S+', string) #Realizamos esta función para conseguir separar la columna en el
#caso de los investigadores.
shark3['Researchers'] = shark3['Investigator or Source'].apply(sharkre)
shark3['Researchers'].head(30)
```

```python
#Hemos visto que cada Researcher estaba en una lista. Así que lo convertimos a string remplpazandolo todo por ',' y luego las borramos
shark3['Researchers'] = [','.join(map(str, l)) for l in shark3['Researchers']]
shark3['Researchers'].replace(to_replace=r',', value='',regex=True, inplace=True)
```
